```python
import keras
import tensorflow
```

```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

# Creating the CNN

### Initialize CNN

```python
classifier = Sequential()
```

### Convolution Layer

```python
classifier.add(Convolution2D(32, 3, 3, border_mode='same', input_shape=(32, 32, 3), activation='relu'))
```

### Pooling Layer

```python
classifier.add(MaxPooling2D(pool_size = (2, 2), ))
```

### 2nd Convolution and Pooling layer

```python
classifier.add(Convolution2D(32, 3, 3, border_mode='same', activation='relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2), ))
```

### Flattening Layer

```python
classifier.add(Flatten())
```

### Fully Connected Layer

```python
classifier.add(Dense(output_dim = 128, activation='relu'))
```

### Output Layer

```python
classifier.add(Dense(output_dim = 1, activation = 'sigmoid'))
```

### Compile model

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

# Traning the Model

```python
from keras.preprocessing.image import ImageDataGenerator
```

### Preparing the Data

```python
train_datagen = ImageDataGenerator(
                rescale=1./255,
                shear_range=0.2,
                zoom_range=0.2,
                horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

training_set = train_datagen.flow_from_directory(
                'dataset/training_set',
                target_size=(32, 32),
                batch_size=32,
                class_mode='binary')

test_set = train_datagen.flow_from_directory(
                'dataset/test_set',
                target_size=(32, 32),
                batch_size=32,
                class_mode='binary')
```

### Fitting the Model

```python
classifier.fit_generator(
                training_set,
                epochs=1,
                steps_per_epoch=8000,
                validation_data=test_set,
                nb_val_samples=2000)
```

### Save the Model

```python
classifier.save('model.h5')
```

# Prediction

```python
import numpy as np
from keras.preprocessing import image
import matplotlib.pyplot as plt
from keras.models import load_model
```

### Load the model

```python
classifier = load_model('model.h5')
```

### Classify Image

```python
 def predict(file_path):
        plt.imshow(plt.imread(file_path))
        
        test_image = image.load_img(file_path, target_size = (32, 32))
        test_image = image.img_to_array(test_image)
        test_image = np.expand_dims(test_image, axis=0)
        result = classifier.predict(test_image)
        
        index = ['']*len(list(training_set.class_indices.keys()))
        for category in list(training_set.class_indices.keys()):
            index[training_set.class_indices[category]] = category
        
        result = index[int(result[0][0])]
        return "Prediction="+result
```

```python
predict('dataset/single_prediction/cat_or_dog_1.jpg')
```
